# Generalized code for terms
# change initial values and equations.

```python
# import packages:

from collections import defaultdict # easy way to create placeholder of dict with list, then to df

#import matplotlib
#from matplotlib import pyplot as plt

import matplotlib.pyplot as plt
#matplotlib.use('TkAgg')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
# Set Initial values
alpha = 0.0001
q = 10
omega = 0
```

```python
# placeholder
stats = defaultdict(list)
```

```python
for step in range(500):
    stats['q'].append(q) # will start with initial value above
    stats['omega'].append(omega) # same here, adds column to df below

    omega = omega + alpha * (1 - omega)
    beta = alpha / omega
    stats['beta'].append(beta)

    reward = np.random.normal(loc=5, scale=1)
    stats['reward'].append(reward)

    q += beta * (reward - q)

```

```python
df = pd.DataFrame().from_dict(stats)
df.head()
```

```python
fig, ax = plt.subplots(nrows=4, figsize=(15,10))

df.plot(y='reward', ax=ax[0])
df.plot(y='q', ax=ax[1])
df.plot(y='omega', ax=ax[2])
df.plot(y='beta', ax=ax[3])

print('final estimate {}'.format(stats['q'][-1]))

fig.show()
```
